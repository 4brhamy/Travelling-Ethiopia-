```python
from collections import deque

class TravelingEthiopiaSearch:
    def __init__(self, map_graph):
        """
        Initialize the search agent with the map of Ethiopia.
        """
        self.map_graph = map_graph

    def search(self, initial_state, goal_state, strategy):
        """
        A wrapper method to switch between BFS and DFS strategies.
        """
        if strategy == "BFS":
            return self.bfs(initial_state, goal_state)
        elif strategy == "DFS":
            return self.dfs(initial_state, goal_state)
        else:
            return None

    def bfs(self, start_node, goal_node):
        """
        Breadth-First Search using a Queue.
        """
        # Frontier: A Queue to store paths. 
        # Structure: (current_node, full_path_to_reach_it)
        frontier = deque([(start_node, [start_node])])
        
        # Explored: A set to keep track of visited cities to avoid loops
        explored = set()

        while frontier:
            # Dequeue from the LEFT (FIFO behavior)
            current_node, path = frontier.popleft()

            # 1. Goal Test
            if current_node == goal_node:
                return path

            # 2. Expand Neighbors
            if current_node not in explored:
                explored.add(current_node)
                
                neighbors = self.map_graph.get(current_node, [])
                for neighbor in neighbors:
                    if neighbor not in explored:
                        new_path = path + [neighbor]
                        frontier.append((neighbor, new_path))
        
        return None # Return None if no path found

    def dfs(self, start_node, goal_node):
        """
        Depth-First Search using a Stack.
        """
        # Frontier: A Stack to store paths.
        frontier = [(start_node, [start_node])]
        explored = set()

        while frontier:
            # Pop from the RIGHT (LIFO behavior)
            current_node, path = frontier.pop()

            if current_node == goal_node:
                return path

            if current_node not in explored:
                explored.add(current_node)
                
                neighbors = self.map_graph.get(current_node, [])
                for neighbor in neighbors:
                    if neighbor not in explored:
                        new_path = path + [neighbor]
                        frontier.append((neighbor, new_path))
        
        return None
```

```python
# Initialize the agent
agent = TravelingEthiopiaSearch(ethiopia_map)

# Test BFS
print("--- BFS Strategy ---")
path_bfs = agent.search("Addis Ababa", "Lalibela", "BFS")
print(f"Path found: {path_bfs}")
print(f"Number of stops: {len(path_bfs)}")

# Test DFS
print("\n--- DFS Strategy ---")
path_dfs = agent.search("Addis Ababa", "Lalibela", "DFS")
print(f"Path found: {path_dfs}")
print(f"Number of stops: {len(path_dfs)}")
```

```python
import heapq

def uniform_cost_search(graph, start, goal):
    # Priority Queue stores tuples of: (cumulative_cost, current_node, path)
    # We start with cost 0 at the start node
    priority_queue = [(0, start, [start])]
    
    # To keep track of visited nodes and the cost to reach them 
    # (to avoid processing more expensive paths to the same node)
    visited_costs = {start: 0}

    while priority_queue:
        # Pop the path with the LOWEST cumulative cost
        current_cost, current_node, path = heapq.heappop(priority_queue)

        # If we reached the goal, return the path and cost
        if current_node == goal:
            return path, current_cost

        # Explore neighbors
        if current_node in graph:
            for neighbor, distance in graph[current_node].items():
                new_cost = current_cost + distance
                
                # If neighbor hasn't been visited or we found a cheaper way to it
                if neighbor not in visited_costs or new_cost < visited_costs[neighbor]:
                    visited_costs[neighbor] = new_cost
                    heapq.heappush(priority_queue, (new_cost, neighbor, path + [neighbor]))

    return None, 0  # Path not found
```
